```python
#importing libraies

import pandas as pd
import numpy as np 
from pandas_datareader import data
```

```python
#Extracting Data from yahoo finance by pandas datareader library.

df=data.DataReader(['HDFCBANK.NS','COROMANDEL.NS','ICICIBANK.NS','BAJAJ-AUTO.NS', 'BHARTIARTL.NS','ICICIGI.NS','IGL.NS','TCS.NS','KOTAKBANK.NS','HDFCLIFE.NS','AXISBANK.NS','ATUL.NS','ASIANPAINT.NS','DMART.NS','BAJAJFINSV.NS','GODREJPROP.NS','BAJFINANCE.NS','ALKEM.NS','ULTRACEMCO.NS','SHREECEM.NS','MUTHOOTFIN.NS','BRITANNIA.NS','VMART.NS','DIVISLAB.NS','DRREDDY.NS'],'yahoo',start='2016/01/01',end='2020/01/01')
```

```python
#Seperate out only useful columns from data

mf_data = df['Adj Close']
mf_data.head()
```

```python
#For checking NAN values if 50% data of a column is null then drop it.

mf_data.info()
```

```python
#droping newly register companies from our data.
mf_data = mf_data.drop(['DMART.NS'],axis=1)
mf_data = mf_data.drop(['ICICIGI.NS'],axis=1)
mf_data = mf_data.drop(['HDFCLIFE.NS'],axis=1)
```

```python
#drop NAN values.  
mf_data = mf_data.dropna()
mf_data
```

```python
#cal. annual return of each asset in fund each year.

j = 1 
annual_return = []
for i in range(4):
    yr1 = mf_data.iloc[j]
    yr1_last = mf_data.iloc[j+237]
    annual_return.append((yr1_last - yr1)/yr1)
    j = j + 237
annual_return        
```

```python
#Averaging out the return to make the annual expected return.

ziped_list = zip(annual_return[0],annual_return[1],annual_return[2],annual_return[3])
sum1 = [x+y+z+w for (x,y,z,w) in ziped_list]
sum1 = [x/4 for x in sum1]
Ex_Return = sum1
Ex_Return
```

```python
#Covariance Matrix.
cov_matrix = mf_data.pct_change().apply(lambda x: np.log(1+x)).cov()
cov_matrix
```

```python
p_return = []
p_vol = []
p_weights = []
num_assets = len(mf_data.columns)
num_portfolios = 5000
```

```python
#Appending annual return, volatility of all 5000 random weights portfolios. 
for i in range(num_portfolios):
    weight = np.random.random(num_assets)
    weight = weight/np.sum(weight)
    p_weights.append(weight)
    ex_return = np.dot(weight,Ex_Return)
    p_return.append(ex_return)
    volitality = cov_matrix.mul(weight,axis=0).mul(weight,axis=1).sum().sum()
    sd = np.sqrt(volitality)
    ann_sd = sd*np.sqrt(238)
    p_vol.append(ann_sd)
```

```python
porfolios_data = {'Returns':p_return,'Volitality':p_vol}
```

```python
porfolios_data = pd.DataFrame(porfolios_data)
porfolios_data.head()
```

```python
for cnt,symbol in enumerate(mf_data.columns.tolist()):
    porfolios_data[symbol+ ' weights'] = [w[cnt] for w in p_weights]
```

```python
#All 5000 portfolios with returns, volitality, each asset weight
porfolios_data
```

```python
from matplotlib import pyplot as plt
```

```python
#plotting all portfolios to visualize our efficent frontier 
porfolios_data.plot.scatter(x='Volitality',y='Returns',grid = True)
plt.gcf().set_size_inches((13, 7))
plt.show()
```

```python
#Minimum variance portfolio with return, volitality and weights.
min_var_por = porfolios_data.iloc[porfolios_data['Volitality'].idxmin()]
min_var_por
```

```python
#Green point showing min. variance portfolios. 

porfolios_data.plot.scatter(x='Volitality',y='Returns',grid = True)
plt.scatter(x = min_var_por[1],y = min_var_por[0],color="green")
plt.gcf().set_size_inches((13, 7))
```

```python
#Return, volatility and weights of optimal risky portfolios. 
Rf = 0.05
Opt_risky_por = porfolios_data.iloc[((porfolios_data['Returns']-Rf)/porfolios_data['Volitality']).idxmax()]
Opt_risky_por
```

```python
#Maximum sharpe Ratio. 
Max_Sharpe_Ratio = (Opt_risky_por[0]-Rf)/Opt_risky_por[1]
Max_Sharpe_Ratio
```

```python
#Black point showing optimal risky portfolios.

porfolios_data.plot.scatter(x='Volitality',y='Returns',grid = True)
plt.scatter(x = min_var_por[1],y = min_var_por[0],color="green",s=50)
plt.scatter(x = Opt_risky_por[1],y = Opt_risky_por[0],color="Black",s=50)
plt.gcf().set_size_inches((13, 7))
```

```python
cal_x = []
cal_y = []
for er in np.linspace(Rf,Opt_risky_por[0],20):
    std_port_rf_optimal = ((er-Rf)/(Opt_risky_por[0]-Rf))*Opt_risky_por[1]
    cal_x.append(std_port_rf_optimal)
    cal_y.append(er)
```

```python
#Plotting capital allocation line.(CAL)
porfolios_data.plot.scatter(x='Volitality',y='Returns',grid = True)
plt.scatter(x = min_var_por[1],y = min_var_por[0],color="green",s=50)
plt.scatter(x = Opt_risky_por[1],y = Opt_risky_por[  0],color="Black",s=50)
plt.plot(cal_x,cal_y)
plt.gcf().set_size_inches((13, 7))
```
